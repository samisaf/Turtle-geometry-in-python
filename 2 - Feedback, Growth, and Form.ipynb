```python
from turtle import *
from math import sin, cos, pi
import random
%matplotlib inline
```

```python
class AnimalTurtle(Turtle):
    def __init__(self, terrarium, boundary = None, color = 'navy'):
        super().__init__(terrarium, color)
        # a boundary is a tuple with the maxmium x and y beyond which the turtle can't move
        self.boundary = boundary
        
    def forward(self, distance):
        degree = pi / 180.0 # conversion between degress and radians
        newx = self.pos.x + distance * cos(self.angle * degree)
        newy = self.pos.y + distance * sin(self.angle * degree)
        if (abs(newx), abs(newy)) > self.boundary: 
            return self.stuck() 
        else:
            return super().forward(distance)
        
    def stuck(self):
        #print("stuck")
        pass
        
    def randmove(self, d1, d2, a1, a2):
        distance = random.randint(d1, d2)
        angle = random.randint(a1, a2)
        self.left(angle)
        return self.forward(distance)

class EdgyTurtle(AnimalTurtle):    
    def wriggle(self):
        self.right(1)
        return self.forward(1)
    
    def stuck(self):
        return self.wriggle()
```

## 2.1 The Turtle as Animal

```python
with Terrarium() as terra:
    cat = AnimalTurtle(terra, (3, 4))
    cat.forward(2).right(90).forward(2).forward(1).left(90).forward(1).forward(1).forward(1)
```

```python
with Terrarium() as terra:
    cat = AnimalTurtle(terra, (3, 4))
    for _ in range(100): cat.randmove(1, 1, 0, 90)
```

```python
with Terrarium() as terra:
    cat = EdgyTurtle(terra, (4, 4))
    for _ in range(100): cat.forward(1).right(1)
```
