## Turtle Graphics
- The turtle(TURTLE) is a little creature that exists in a mathematical plan (TERRARIUM) on a computer display. It can resond to a few commands: FORWARD, BACKWARD, RIGHT, LEFT
- You can trace the path of the turtle by asking it to hold a pen down while moving. The command is: PENDOWN
- We can define procedures that combines simple commands to construct more complex shapes.

```python
# Initializing commands
from turtle import Turtle, Terrarium
%matplotlib inline
```

```python
# simple example
terra = Terrarium()
t1 = Turtle(terra)
t1.forward(100); t1.right(60); t1.back(50)
terra.rescale()
```

```python
# same above example, but using chaining procedures instead. 
# Note that chaining works for basic commands, but might not work with user defined functions
terra = Terrarium()
t1 = Turtle(terra)
t1.forward(100).right(60).back(50)
terra.rescale()
```

```python
# drawing a square
with Terrarium() as terra:
    t1 = Turtle(terra)
    t1.forward(100); t1.right(90)
    t1.forward(100); t1.right(90)
    t1.forward(100); t1.right(90)
    t1.forward(100); t1.right(90)
```

```python
# drawing a square by chaining commands
with Terrarium() as terra:
    Turtle(terra).forward(100).right(90).forward(100).right(90).forward(100).right(90).forward(100).right(90)
```

```python
# drawing a square
with Terrarium() as terra:
    t1 = Turtle(terra)
    for _ in range(4): t1.forward(100); t1.right(90)
```

```python
# drawing a square by defing a procedure

def square(self, length):
    for _ in range(4):
        self.forward(length)
        self.right(90)

Turtle.square = square

with Terrarium() as terra:
    t1 = Turtle(terra)
    t1.square(100)
```

```python
# drawing a triangle, note using lambda, and chaining commands

Turtle.triangle = lambda self, length: [self.forward(length).right(120) for _ in range(3)]

with Terrarium() as terra:
    t1 = Turtle(terra)
    t1.triangle(100)
```

```python
# drawing a simple house

length = 10
with Terrarium() as terra:
    t1 = Turtle(terra)
    t1.square(length)
    t1.forward(length)
    t1.right(30)
    t1.triangle(length)
    
```

```python
# drawing a pattern

def thing(self):
    self.forward(100).right(90).forward(100).right(90)
    self.forward(50).right(90).forward(50).right(90)
    self.forward(100).right(90).forward(25).right(90)
    self.forward(25).right(90).forward(50)
    
Turtle.thing = thing

with Terrarium() as terra:
    Turtle(terra).thing()

with Terrarium() as terra:
    t = Turtle(terra)
    for _ in range(4):
        t.thing()

with Terrarium() as terra:
    t = Turtle(terra)
    for _ in range(10):
        t.thing()
        t.right(50)
        t.forward(50)
        
with Terrarium() as terra:
    t = Turtle(terra)
    for _ in range(10):
        t.thing()
        t.left(45)
        t.forward(100)
```

```python
# drawing a circle
with Terrarium() as terra:
    t = Turtle(terra)
    for _ in range(360):
        t.forward(1).right(1)
```

```python
# drawing arcs

Turtle.arcr = lambda self, r, deg: [self.forward(r).right(1) for _ in range(deg)]
Turtle.arcl = lambda self, r, deg: [self.forward(r).left(1) for _ in range(deg)]

def petal(self, size):
    self.arcr(size, 60)
    self.right(120)
    self.arcr(size, 60)
    self.right(120)

Turtle.petal = petal

length = 10
with Terrarium() as terra:
    t = Turtle(terra)
    for _ in range(6):
        t.petal(length)
        t.right(60)
```

```python
# drawing sun

def ray(self, r):
    for _ in range(2):
        self.arcl(r, 90)
        self.arcr(r, 90)
        
Turtle.ray = ray

length = 10
with Terrarium() as terra:
    t = Turtle(terra)
    for _ in range(9):
        t.ray(length)
        t.right(160)
```

```python
length = 10
with Terrarium(animate = True) as terra:
    t = Turtle(terra)
    for _ in range(9):
        t.ray(length)
        t.right(160)
```

## Some Simple Turtle Programs

```python
def poly(turtle, side, angle):
    for _ in range(10):
        turtle.forward(side)
        turtle.right(angle)
    return turtle

Turtle.poly = poly

with Terrarium() as terra:
    Turtle(terra).poly(10, 72)
    Turtle(terra, 'red').poly(10, 144)

with Terrarium() as terra:
    Turtle(terra, 'orange').poly(10, 108)
```

```python
def polyspi(turtle, side, angle, inc=1, recdepth=50):
    for _ in range(recdepth):
        turtle.forward(side)
        turtle.right(angle)
        side = side + inc
    return turtle

Turtle.polyspi = polyspi

with Terrarium() as terra:
    Turtle(terra).polyspi(10, 95)
    
with Terrarium() as terra:
    Turtle(terra).polyspi(10, 90)

with Terrarium() as terra:
    Turtle(terra).polyspi(10, 120)

with Terrarium() as terra:
    Turtle(terra).polyspi(10, 117)
```
